```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
```

Library
1. pandas untuk manipulasi data.
2. numpy untuk operasi numerik.
3. mlxtend.frequent_patterns untuk algoritma apriori dan aturan asosiasi.

Proses:
1. apriori: Mengidentifikasi itemset yang sering muncul dalam dataset transaksi.
2. association_rules: Membuat aturan asosiasi dari itemset yang sering muncul, yang membantu menemukan hubungan antar item dalam dataset.

```python
df = pd.read_csv('GroceryStoreDataSet.csv', names = ['products'], sep = ',')
df.head()
```

1. pd.read_csv digunakan untuk membaca file CSV bernama 'GroceryStoreDataSet.csv'.
2. names=['products'] menambahkan nama kolom 'products' pada dataframe yang dihasilkan.
3. sep=',' menentukan bahwa pemisah (delimiter) dalam file CSV adalah koma.

```python
df.shape
```

atribut dari objek NumPy array atau Pandas DataFrame yang memberikan informasi tentang dimensi data. Ia mengembalikan tuple yang menunjukkan jumlah baris dan kolom dalam data.

```python
data = list(df["products"].apply(lambda x:x.split(",") ))
data
```

1. df["products"] mengakses kolom "products" dari dataframe df.
2. .apply(lambda x: x.split(",")) menerapkan fungsi lambda yang membagi setiap string x (representasi transaksi) berdasarkan koma menjadi sebuah list.
3. list() mengubah hasil dari operasi sebelumnya menjadi sebuah list Python.

**Apriori**

```python
#Let's transform the list, with one-hot encoding
from mlxtend.preprocessing import TransactionEncoder
a = TransactionEncoder()
a_data = a.fit(data).transform(data)
df = pd.DataFrame(a_data,columns=a.columns_)
df = df.replace(True,1)
df = df.replace(False,0)
df
```

**Import dan Inisialisasi TransactionEncoder**

1. TransactionEncoder dari mlxtend.preprocessing digunakan untuk mengkodekan data transaksi menjadi matriks one-hot encoding.
**Transformasi Data**

1. a.fit(data) mengadaptasi TransactionEncoder ke data transaksi data.
2. .transform(data) mengubah data transaksi data menjadi matriks one-hot encoding.
**Konversi ke DataFrame Pandas**

1. pd.DataFrame(a_data, columns=a.columns_) mengubah matriks a_data menjadi dataframe Pandas.
2. columns=a.columns_ menggunakan nama kolom dari TransactionEncoder sebagai nama kolom dalam dataframe.
**Mengganti True dan False dengan 1 dan 0**

1. df.replace(True, 1) mengganti nilai True (representasi dari one-hot encoding yang tidak ada dalam transaksi) dengan nilai 1.
2. df.replace(False, 0) mengganti nilai False (representasi dari one-hot encoding yang tidak ada dalam transaksi) dengan nilai 0.


```python
#set a threshold value for the support value and calculate the support value.
df = apriori(df, min_support = 0.2, use_colnames = True, verbose = 1)
df
```

menggunakan fungsi apriori dari pustaka mlxtend.frequent_patterns untuk melakukan analisis frequent itemset (itemset sering muncul) berdasarkan matriks one-hot encoding yang sudah dibuat sebelumnya.



```python
#Let's view our interpretation values using the Associan rule function.
df_ar = association_rules(df, metric = "confidence", min_threshold = 0.6)
df_ar
```

menggunakan fungsi association_rules dari pustaka mlxtend.frequent_patterns untuk mengekstrak aturan asosiasi dari frequent itemset yang sudah dihasilkan sebelumnya.
